Loading the Dataset


```python
import pandas as pd

df = pd.read_csv('student_marks.csv')

df.head()
```

Distribution of average scores with matplotlib

```python
import matplotlib.pyplot as plt
import seaborn as sns

df['Average_Score'] = df.loc[:, 'Test_1':'Test_12'].mean(axis=1)

plt.figure(figsize=(8,5))
sns.histplot(df['Average_Score'], bins=10, kde=True)
plt.title('Distribution of Average Scores')
plt.xlabel('Average Score')
plt.ylabel('Number of Students')
plt.show()
```

Displaying Summary statistics of all the tests

```python
# Summary statistics
print(df.loc[:, 'Test_1':'Test_12'].describe())
print("\nAverage Score Stats:")
print(df['Average_Score'].describe())
```

Boxplot between tests and scores

```python
#Boxplot
plt.figure(figsize=(12,6))
sns.boxplot(data=df.loc[:, 'Test_1':'Test_12'])
plt.title('Score Distribution Across Tests')
plt.xlabel('Tests')
plt.ylabel('Scores')
plt.show()
```

Finding correlation heatmap

```python
#correlation heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df.loc[:, 'Test_1':'Test_12'].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Test Scores')
plt.show()
```

Identifying top 5 students by their average score

```python
top_students = df[['Student_ID', 'Average_Score']].sort_values(by='Average_Score', ascending=False).head(5)
print("Top 5 Students by Average Score:")
print(top_students)
```
