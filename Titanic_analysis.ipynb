Loading the csv file

```python
import pandas as pd
df = pd.read_csv('Titanic.csv')
print(df.head())
```

Cleaning the dataset


```python

df.dropna(inplace = True)
print(df.head())
```

```python

df.dropna(subset=['Age'],inplace=True)
print(df.head())
```

```python


df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Fare'] = df['Fare'].fillna(df['Fare'].median())
print(df.head())
```

```python

df.drop_duplicates(inplace=True)
print(df.head())
```

Exploratory Data Analysis

Scatter plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data=df, x='Age', y='Survived', alpha=0.7, hue='Survived', palette={0: 'red', 1: 'blue'})
plt.xlabel('Age')
plt.ylabel('Survived')
plt.title('Scatter Plot of Age vs. Survival')
plt.yticks([0, 1], ['Not Survived', 'Survived'])  # Optional: Label y-axis for clarity
plt.show()
```

```python
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived', palette={0: 'red', 1: 'blue'}, alpha=0.7)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs. Fare')
plt.show()
```

Boxplot

```python
# Create age bins
df['AgeBin'] = pd.cut(df['Age'], bins=range(0, 101, 10), right=False)

# Box plot of Fare by Age Bin and Survival Status
sns.boxplot(data=df, x='AgeBin', y='Fare', hue='Survived', palette={0: 'red', 1: 'green'})
plt.xlabel('Age Bin')
plt.ylabel('Fare')
plt.title('Box Plot of Fare by Age Bin and Survival Status')
plt.xticks(rotation=45)
plt.legend(title='Survived', labels=['Not Survived', 'Survived'])
plt.show()
```

```python
sns.boxplot(data=df, x='Pclass', y='Age', palette='Set2')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.title('Box Plot of Age by Passenger Class')
plt.show()
```

Violin Plot


```python
# Create age bins
df['AgeBin'] = pd.cut(df['Age'], bins=range(0, 101, 10), right=False)

# Violin plot of Fare by Age Bin and Survival Status
sns.violinplot(data=df, x='AgeBin', y='Fare', hue='Survived', palette={0: 'red', 1: 'green'}, split=True)
plt.xlabel('Age Bin')
plt.ylabel('Fare')
plt.title('Violin Plot of Fare by Age Bin and Survival Status')
plt.xticks(rotation=45)
plt.legend(title='Survived', labels=['Not Survived', 'Survived'])
plt.show()
```

```python
sns.violinplot(data=df, x='Pclass', y='Age', palette='Set2')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.title('Violin Plot of Age by Passenger Class')
plt.show()
```

Logistic Regression Model

```python
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df['Embarked'] = df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Selecting features and target
features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']
X = df[features]
y = df['Survived']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# Model
model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```
